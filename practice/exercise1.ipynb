```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
days = np.linspace(0,60,1000)
```

```python
EnergyM1 = 1 * days
EnergyM2 = 4 * (days - 30)
```

```python
fig, ax = plt.subplots()
plt.title('Energy of a Jil design')
plt.xlabel('month')
plt.ylabel('Energy')
ax.set_xlim([0,60])
ax.set_ylim([0,100])
ax.plot(days, EnergyM1, label='EnergyM1',c='green')
ax.plot(days, EnergyM2, label='EnergyM2',c='red')
plt.legend()
plt.axvline(x=40, color='black', linestyle='--')  == plt.axhline(y=40, color='black', linestyle='--')
plt.xticks([0,30],['april','may'])
```
