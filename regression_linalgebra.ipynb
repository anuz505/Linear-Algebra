```python
import torch
```

```python
x1 = torch.tensor([0,1,2,3,4,5,6,7],dtype=torch.float32)
x1
```

```python
y = torch.tensor([1.86,1.31,.62,.33,.09,-.67,-1.23,-1.37],dtype=torch.float32)
y
```

```python
import matplotlib.pyplot as plt
```

```python
title = 'Clinical Trial'
xlabel = 'Drug dosage (mL)'
ylabel = 'Forgetfulness'
```

```python
fig, ax = plt.subplots()
plt.title(title)
plt.xlabel(xlabel)
plt.ylabel(ylabel)
_ = ax.scatter(x1, y)
```

```python
x0 = torch.ones_like(x1)
x0
```

```python
X = torch.stack((x0, x1), dim=1)
X
```

```python
w = torch.matmul(torch.linalg.pinv(X),y)
w
```

```python
# b0 = w[0] # y-intercept
b0 = w[0].item()
b0
```

```python
# b1 = w[1] # slope
b1 = w[1].item()
b1
```

```python
fig, ax = plt.subplots()
plt.title(title)
plt.xlabel(xlabel)
plt.ylabel(ylabel)
_ = ax.scatter(x1, y)

x_min, x_max = ax.get_xlim()
Y_xmin = b0 + b1 * x_min
Y_xmax = b0 + b1 * x_max
ax.plot([x_min, x_max], [Y_xmin, Y_xmax], color='r')
```

### And this is how we do regression using Linear algebra only.
