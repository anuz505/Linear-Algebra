```python
import numpy as np
import matplotlib.pyplot as plt

# Define x values from 0 to 2pi with small increments
x = np.linspace(0, 2 * np.pi, 10000)

# Define y = sec(x), which is 1/cos(x)
y = 1 / np.cos(x)

a = np.cos(x)

# Create a mask to handle asymptotes (where cos(x) is zero)
y[np.abs(np.cos(x)) < 0.1] = np.inf  # Set points where cos(x) is close to zero to infinity (asymptotes)

# Plot the sec(x) function
plt.plot(x, y, label=r'$y = \sec(x)$')
plt.plot(x, a, label=r'$y = \cos(x)$')

# Add vertical asymptotes at x = pi/2 and 3pi/2
plt.axvline(np.pi / 2, color='r', linestyle='--', label=r'$x = \frac{\pi}{2}$ (asymptote)')
plt.axvline(3 * np.pi / 2, color='r', linestyle='--', label=r'$x = \frac{3\pi}{2}$ (asymptote)')

# Set plot limits for better visualization
plt.xlim(0, 2 * np.pi)
plt.ylim(-3, 3)

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title(r'Plot of $y = \sec(x)$ from $x = 0$ to $x = 2\pi$')

# Add grid, legend, and show the plot
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Functions and Their Graphs

## Import Necessary Libraries
import numpy as np
import matplotlib.pyplot as plt

# Set up the figure size for all plots
plt.rcParams['figure.figsize'] = (10, 6)

# 1. Linear Function
def linear_function(x):
    return 2 * x + 1

x = np.linspace(-10, 10, 100)
y = linear_function(x)

plt.plot(x, y, label='y = 2x + 1', color='blue')
plt.title('Linear Function')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()

# 2. Quadratic Function
def quadratic_function(x):
    return x**2 - 4*x + 3

y = quadratic_function(x)

plt.plot(x, y, label='y = x^2 - 4x + 3', color='green')
plt.title('Quadratic Function')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()

# 3. Cubic Function
def cubic_function(x):
    return x**3 - 3*x**2 + 2

y = cubic_function(x)

plt.plot(x, y, label='y = x^3 - 3x^2 + 2', color='red')
plt.title('Cubic Function')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()

# 4. Exponential Function
def exponential_function(x):
    return np.exp(x)

y = exponential_function(x)

plt.plot(x, y, label='y = e^x', color='orange')
plt.title('Exponential Function')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()

# 5. Logarithmic Function
def logarithmic_function(x):
    return np.log(x)

x_positive = np.linspace(0.1, 10, 100)
y = logarithmic_function(x_positive)

plt.plot(x_positive, y, label='y = ln(x)', color='purple')
plt.title('Logarithmic Function')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()

# 6. Trigonometric Functions
x = np.linspace(-2 * np.pi, 2 * np.pi, 100)

y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin, label='y = sin(x)', color='cyan')
plt.plot(x, y_cos, label='y = cos(x)', color='magenta')
plt.title('Trigonometric Functions: Sine and Cosine')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()

# 7. Absolute Value Function
def absolute_value_function(x):
    return np.abs(x)

y = absolute_value_function(x)

plt.plot(x, y, label='y = |x|', color='brown')
plt.title('Absolute Value Function')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()

# 8. Step Function
def step_function(x):
    return np.where(x >= 0, 1, 0)

y = step_function(x)

plt.step(x, y, label='y = step(x)', color='black')
plt.title('Step Function')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()
```
